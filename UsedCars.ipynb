```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "vehicles.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
if "price" in df.columns:
    df_filtered = df[(df["price"] > 100) & (df["price"] <= 100000)]
else:
    raise ValueError("The dataset does not contain a 'price' column")
```

```python
if "manufacturer" in df_filtered.columns:
    manufacturer_stats = df_filtered.groupby("manufacturer")["price"].agg(["mean", "median", "min", "max", "std", "count"])
    manufacturer_stats_df = manufacturer_stats.reset_index()
    print(manufacturer_stats_df.head())
else:
    raise ValueError("The dataset does not contain a 'manufacturer' column")
```

```python
if "year" in df_filtered.columns and "price" in df_filtered.columns:
    plt.figure(figsize =(8,5))
    plt.scatter(df_filtered["year"] , df_filtered["price"] , alpha=0.2)
    plt.xlabel("Year")
    plt.ylabel("Price ($)")
    plt.title("Car Price vs Year")
    plt.grid(True)
    plt.show()
else:
    raise ValueError("The dataset does not contain 'year' or 'price' column")
```

```python
if "odometer" in df_filtered.columns and "price" in df_filtered.columns:
    plt.figure(figsize =(8,5))
    plt.scatter(df_filtered["odometer"] , df_filtered["price"] , alpha=0.2 , color="red")
    plt.xlabel("Odometer (miles)")
    plt.ylabel("Price ($)")
    plt.title("Car Price vs Odometer")
    plt.grid(True)
    plt.show()
else:
    raise ValueError("The dataset does not contain 'odometer' or 'price' column")
```
